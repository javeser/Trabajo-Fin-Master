```python
import uproot
import numpy as np
import matplotlib.pyplot as plt

def Jpsi_variables():
    # Abrir el archivo ROOT y obtener el Ã¡rbol
    filename = "Lb2_JpsiLambda_vtx.root"
    tree = uproot.open(filename)["Hlt2DetachedJpsiTuple/Tuple"]

    # Obtener los nombres de las ramas del Ã¡rbol
    branch_names = tree.keys()

    # Definir variables para almacenar los nombres filtrados
    filtered_names = []

    # Loop sobre todas las ramas
    for name in branch_names:
        # Verificar si la variable contiene ciertas palabras clave
        if any(keyword in name for keyword in ["Jpsi_END_VZ", "Jpsi_END_VX", "Jpsi_END_VY", "Jpsi_END_VRHO", "Jpsi_BPVCORRM", "Jpsi_BPVDIRA", "Jpsi_BPVETA", "Jpsi_BPVIPCHI2", "Jpsi_BPVFDCHI2", "Jpsi_BPVLTIME", "Jpsi_BPVFD", "Jpsi_BPVFDIRX", "Jpsi_BVPFDIRY", "Jpsi_BPVDIRZ", "Jpsi_BPVFDVECX", "Jpsi_BPVFDVECY", "Jpsi_BPVFDVECZ", "Jpsi_BPVIP", "Jpsi_BPVVRHO", "Jpsi_BPVVDX", "Jpsi_BPVVDY", "Jpsi_BPVVDZ", "Jpsi_COMB_MASS", "Jpsi_MAXDOCA", "Jpsi_MAXDOCACHI2", "Jpsi_MAXSDOCA", "Jpsi_MAXSDOCACHI2", "Jpsi_MASS", "Jpsi_P", "Jpsi_PT", "Jpsi_PX", "Jpsi_PY", "Jpsi_PZ", "Jpsi_E", "Jpsi_ETA", "Jpsi_PHI", "Jpsi_CHI2DOF", "Jpsi_CHI2", "Jpsi_FOURMOMENTUME", "Jpsi_FOURMOMENTUMX", "Jpsi_FOURMOMENTUMY", "Jpsi_FOURMOMENTUMZ"]):
            filtered_names.append(name)

    # Determinar el nÃºmero de filas y columnas para el lienzo
    rows = 4
    cols = 3

    # Variables contadoras para el lienzo y el histograma
    canv_count = 0
    hist_count = 0

    # Loop sobre los nombres de las variables filtradas
    for i, var in enumerate(filtered_names):
        # Obtener los datos de la variable de la rama especÃ­fica
        data = tree[var].array()

        # Obtener los datos para seÃ±al y fondo
        bkgs = data[tree["Jpsi_BKGCAT"].array() != 0]
        signals = data[tree["Jpsi_BKGCAT"].array() == 0]

        # Crear un nuevo lienzo cada cols*rows histogramas
        if hist_count % (rows * cols) == 0:
            canv_count += 1
            plt.figure(figsize=(16, 15))

        # Trazar histogramas
        plt.subplot(rows, cols, hist_count % (rows * cols) + 1)
        plt.hist(bkgs, bins=100, color='red', alpha=0.5, label='Background', histtype='step', density=True)
        plt.hist(signals, bins=100, color='blue', alpha=0.5, label='Signal', histtype='step', density=True)
        plt.xlabel(var)
        plt.ylabel('Entries')
        plt.legend()

        # Establecer escala logarÃ­tmica para ciertas variables
        if var in ["Jpsi_END_VRHO", "Jpsi_BPVCORRM", "Jpsi_BPVDIRA", "Jpsi_BPVIPCHI2", "Jpsi_BPVFD", "Jpsi_BPVIP", "Jpsi_BPVVDZ", "Jpsi_MAXSDOCACHI2", "Jpsi_MAXDOCACHI2", "Jpsi_CHI2DOF", "Jpsi_CHI2"]:
            plt.xscale('log')

        hist_count += 1

    plt.show()

Jpsi_variables()
```
